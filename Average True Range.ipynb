### We Calculate the ATR by calculating the following things first

#### Period High minus (-) Current Period Low
#### The Absolute Value (abs) of the Current Period High minus (-) The Previous Period Close
#### The Absolute Value (abs) of the Current Period Low minus (-) The Previous Period Close
#### true range = max[(high - low), abs(high - previous close), abs (low - previous close)]
More Detailed Explaination - https://www.tradingview.com/scripts/averagetruerange/

```python
# The ATR Function - 
```

```python
def ATR(DF , n = 14):
    df = DF.copy() #We do not want to affect the original Dataframe.
    df["H-L"] = df["High"] - df["Low"]
    df["H-PC"] = df["High"] - df["Adj Close"].shift(1)
    df["L-PC"] = df["Low"] - df["Adj Close"].shift(1)
    df["True_Range"] = df[["H-L" , "H-PC" , "L-PC"]].max(axis = 1 , skipna = False)
    df["ATR"] = df["True_Range"].ewm(span = n , min_periods = n).mean()
    return df["ATR"]
```

### Example - 

```python
import yfinance
```

```python
Google = yfinance.download(tickers = "GOOG" , period = "1mo" , interval = "5m")
```
