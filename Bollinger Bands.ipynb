## How to Calculate Bollinger Bands in Python
## By Bankej Soni

### We Calculate the following things first - 

### Middle Band – 20 Day Simple Moving Average
### Upper Band – 20 Day Simple Moving Average + (Standard Deviation x 2)
### Lower Band – 20 Day Simple Moving Average - (Standard Deviation x 2)

More Detailed Explaination - https://www.tradingview.com/scripts/bollingerbands/

```python
# Bollinger Bands Function
```

```python
def Boll_Band(DF , n = 20):
    df = DF.copy()
    df["MB"] = df["Adj Close"].rolling(n).mean()
    df["UB"] = df["MB"] + 2*df["Adj Close"].rolling(n).std(ddof = 0)
    df["LB"] = df["MB"] - 2*df["Adj Close"].rolling(n).std(ddof = 0)
    df["Band_width"] = df["UB"]-df["LB"]
    return df[["MB","UB","LB","Band_width"]]
```

### Example

```python
import yfinance
```

```python
Google = yfinance.download(tickers = "GOOG" , period = "1mo" , interval = "5m")
```

```python
Google
```

```python
#Applying the Function
```

```python
Google[["MB","UB","LB","Band_width"]] = Boll_Band(Google)
```
