```python
# The SMA Function (Simple Moving Average)
```

```python
def SMA(DF, n): # where n = length
    df = DF.copy()
    df["SMA"] = df["daily_return"].rolling(window = n).mean()
    return df["SMA"]
```

### Example - 

```python
import yfinance
```

```python
Google = yfinance.download(tickers = "GOOG" , period = "1mo" , interval = "5m")
```

```python
Google
```

```python
Google["daily_return"] = (Google["Adj Close"]/Google["Adj Close"].shift(1)) - 1 # Calculating the Daily Returns
Google
```

```python
Google["SMA"] = SMA(Google , 20)
```
